### Генерация признаков

Для начала объединим данные о покупках для каждого клиента, за один день

````python
def comb_lines(df):
    df_new = df.groupby(['client_id', 'date', 'message_id']) \
               .agg({'quantity': 'sum', 'price': 'sum'}) \
               .reset_index()
    return df_new
````

#### Хэширование message_id

Идентификатор сообщения из рассылки может быть одним из факторов влияющий на покупки, но его необходимо преобразовать в формат, который модель может интерпретировать. Для этого будем использовать функцию хэширования

````python
# Функция для хэширования message_id

def hash_message_id(message_id):
    return int(hashlib.sha256(message_id.encode()).hexdigest(), 16) % 10**8
````

````python
apparel['hashed_message_id'] = apparel['message_id'].apply(hash_message_id)
gifts['hashed_message_id'] = gifts['message_id'].apply(hash_message_id)
marketplace['hashed_message_id'] = marketplace['message_id'].apply(hash_message_id)
````

Функция генерирует хэш-код SHA-256 для строки message_id, затем преобразует его в целое число, используя шестнадцатеричное представление, и ограничивает результат в диапазоне от 0 до 99 999 999

**Вовлечённость клиента**

Нас интересует интервал времени до следующего посещения, лежащий в диапазоне [t<sub>s</sub>,∞] где t<sub>s</sub> это таймаут, которым разделяются сессии. Чем меньше этот интервал, тем более вовлечён клиент. Если предсказанный интервал близок к бесконечности, то клиент возможно не вернётся на *сайт* никогда, и можно считать, что он churned. Хорошо бы использовать прогнозное значение, как **метрику вовлечённости клиента** (т.е. насколько он “churned” или “not churned”). Клиент, который вернётся завтра, очевидно, более вовлечён, чем клиент, который вернётся через полгода. Из этого, генерим признак:
- совершил ли клиент **повторную покупку** в течение 30 дней

#### Признак: разница по времени

Рассчитаем время в днях с последней покупки для каждого пользователя

Время между покупками может помочь ответить на некоторые базовые вопросы любого бизнеса:

- Как часто клиенты совершают покупки?
- Сколько времени в среднем проходит между покупками?
- Клиент активен или является ушедшим?
- Какова доля активных клиентов в клиентской базе?
- Какова доля ушедших клиентов в клиентской базе?
- Какие клиенты с большой вероятностью вскоре прекратят сотрудничество?

Для клиентов, что севершили одну и/или первую покупку поставим заглушку `-1`

````python
# Разница по времени

def time_diff(df):

    delta = (
        df
        .sort_values(by=['client_id', 'date'])
        .groupby('client_id')['date'].diff().dt.days
        .fillna(-1)
    )
    return delta
````

````python
apparel['delta'] = time_diff(apparel)
gifts['delta'] = time_diff(gifts)
marketplace['delta'] = time_diff(marketplace)
````

#### Признак: повторная покупка

Задача прогноза времени до наступления события (в нашем случае - возврат клиента) встречается довольно часто, но теория анализа выживаемости и соответствующие методики прогнозирования сначала были разработаны для медицинских исследований. Событием, наступление которого анализировалось, являлась смерть пациента. Такой анализ часто применяется при сравнении групп пациентов, получивших разные способы лечения: более эффективен тот способ, где пациенты прожили дольше. В нашем случае логика **инвертируется**: хороший клиент это тот, кто вернулся (“умер” с точки зрения survival analysis).

Сгенерируем признак `churned` для всех данных. Если интервал вовлеченности клиента **больше 30 дней** с момента последней покупки, то он **churned**

````python
# Вовлеченность клиента

def customer_involvement(df):

    churned = ((df['delta'] >= 0) & (df['delta'] < 30)).astype(int)
    #churned = (df['delta'] < 30).astype(int)
    return churned
````

````python
apparel['churned'] = customer_involvement(apparel)
apparel['churned'].value_counts(dropna=False)
````

````python
gifts['churned'] = customer_involvement(gifts)
gifts['churned'].value_counts(dropna=False)
````

````python
marketplace['churned'] = customer_involvement(marketplace)
marketplace['churned'].value_counts(dropna=False)
````

#### Признак: день, день недели, скользящее среднее, отстающие значения

**Найдем скользящее среднее для:**

- Количество товара в заказе (quantity): можно определить общий тренд в покупательской способности клиента.

- Цена товара в заказе (price): скользящее среднее цен товаров может помочь в выявлении общего изменения в тенденциях цен или стоимости товаров.

**Максимальный размер отставания:**

- Цена товаров в заказе: отставание здесь может предоставить информацию о динамике изменений цен

````python
def make_features(df, max_lag, rolling_mean_size):

    # День, день недели, месяц
    df['day'] = df['date'].dt.day
    df['dayofweek'] = df['date'].dt.dayofweek
    df['month'] = df['date'].dt.month

    # Максимальный размер отставания
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['price'].shift(lag)

    # Скользящее среднее
    df['rolling_quantity'] = df['quantity'].shift().rolling(rolling_mean_size).mean()
    df['rolling_price'] = df['price'].shift().rolling(rolling_mean_size).mean()

    return df
````

````python
make_features(apparel, 3, 3).dropna(inplace=True)
make_features(gifts, 3, 3).dropna(inplace=True)
make_features(marketplace, 3, 3).dropna(inplace=True)
````

### График сезонности

Для графика сезонности сделаем срез данных за последний месяц, так он будет более информативен

````python
# График сезонности
def seas_decom(df):

    decomposed = seasonal_decompose(df)

    plt.figure(figsize=(8, 10))
    plt.subplot(311)

    decomposed.trend.plot(ax=plt.gca())
    plt.title('Trend')
    plt.xlabel('Месяц')
    plt.subplot(312)

    decomposed.seasonal[:30].plot(ax=plt.gca())
    plt.title('Seasonality')
    plt.xlabel('Месяц')
    plt.subplot(313)

    decomposed.resid.plot(ax=plt.gca())
    plt.title('Residuals')
    plt.xlabel('Месяц')
    plt.tight_layout()
````

````python
# График стандартного отклонения
def std_plt(df):

    df['mean'] = df_m['price'].rolling(15).mean()
    df['std'] = df_m['price'].rolling(15).std()
    df.plot(figsize=(9, 4))

    plt.title('График стандартного отклонения по скользящему окну')
    plt.xlabel('Месяц')
    plt.show()
````

````python
# График разностей временного ряда
def time_plt(df):
    df = df - df.shift()

    df['mean'] = df['price'].rolling(15).mean()
    df['std'] = df['price'].rolling(15).std()
    df.plot(figsize=(9, 4))

    plt.title('График распределения разностей временного ряда')
    plt.xlabel('Месяц')
    plt.show()
````

###  Функции

````python
# Важность признаков
def feature_importance(model, X):

    importances = model['classifier'].feature_importances_
    df = pd.DataFrame({'Feature': X.columns, 'Importance': importances}
                     ).sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(13, 6))
    sns.barplot(x='Importance', y='Feature', data=df, palette='Blues')
    plt.title('Feature Importance')
    plt.show()
````

````python
# Матрица ошибок
def error_matrix(model, X, y):

    y_pred = model.predict(X)
    cm = confusion_matrix(y, y_pred)

    plt.figure(figsize = (8, 6))
    plt.title('Error matrix')

    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')

    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
````

### Обучение

#### Тренировка и подготовка моделей

````python
%%time

time_cv = TimeSeriesSplit(n_splits=5)

pipeline = Pipeline([
    ('standard_scaler', StandardScaler()),
    ('classifier', CatBoostClassifier(verbose=0))
])

params = {
    'classifier__max_depth': [4, 6, 8],
    'classifier__learning_rate': [0.01, 0.05, 0.1]
}

grid_search = GridSearchCV(pipeline,
                           params,
                           scoring='f1',
                           cv=time_cv,
                           n_jobs=-1,
                           error_score='raise')

grid_search.fit(X_train, y_train)

print(f'best_score: {grid_search.best_score_}')
print(f'best_params: {grid_search.best_params_}')
````

Вручную зададим веса кслассов: для положительного класс 2, иначе 1

````python
%%time

time_cv = TimeSeriesSplit(n_splits=5)


pipeline = Pipeline([
    ('standard_scaler', StandardScaler()),
    ('model', GradientBoostingClassifier())
])

params = {
    'model__n_estimators': [50, 100, 200],
    'model__learning_rate': [0.01, 0.1, 0.2],
    'model__max_depth': [3, 5, 7]
}

grid = GridSearchCV(pipeline,
                    params,
                    scoring='f1',
                    cv=time_cv,
                    n_jobs=-1,
                    error_score='raise')

grid.fit(X_train, y_train, model__sample_weight=np.where(y_train == 1, 2, 1))

print(f'Best F1-score: {grid.best_score_}')
print(f'Best parameters: {grid.best_params_}')
````

````python
feature_importance(grid_search.best_estimator_, X_train)
````

````python
error_matrix(grid_search, X_train, y_train)
````

Модель более успешно предсказывает отрицательные результаты, чем положительные.

#### Тестирование лучшей модели

````python
y_pred = grid.best_estimator_.predict(X_test)
f1_value = f1_score(y_test, y_pred)

print("F1_score на тестовом наборе данных:", f1_value)
````

````python
report = classification_report(y_test, y_pred)
print(report)
````

````python
probabilities_valid = grid.predict_proba(X_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, probabilities_valid)

plt.figure()
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')

plt.plot([0, 1], [0, 1], linestyle='--')

auc_roc = roc_auc_score(y_test, probabilities_valid)

print('Best Grid:', grid.best_params_)
print("AUC-ROC:", auc_roc)
````
